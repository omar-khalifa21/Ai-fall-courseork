## Now we add some features 

```python
from datetime import datetime
# ---- Total spending ----
df["total_spending"] = (
    df["spend_wine"] +
    df["spend_fruits"] +
    df["spend_meat"] +
    df["spend_fish"] +
    df["spend_sweets"] +
    df["spend_gold"]
)

# ---- Family size ----
# married = 2 adults, else = 1 adult
df["family_size"] = df["num_children"] + df["num_teenagers"] + df["marital_status"].apply(
    lambda x: 2 if x == "Married" else 1
)

# ---- Number of accepted campaigns ----
campaign_cols = [
    "accepted_campaign_1",
    "accepted_campaign_2",
    "accepted_campaign_3",
    "accepted_campaign_4",
    "accepted_campaign_5",
    "accepted_last_campaign"
]
df["accepted_campaign_count"] = df[campaign_cols].sum(axis=1)

# ---- Number of purchases ----
df["num_purchases"] = (
    df["num_web_purchases"] +
    df["num_catalog_purchases"] +
    df["num_store_purchases"] +
    df["num_discount_purchases"]
)

#----days active-------

df["signup_date"] = pd.to_datetime(df["signup_date"], dayfirst=True, errors='coerce')

# Drop rows that still failed to parse
df = df.dropna(subset=["signup_date"])

# Compute customer_days_active
df["customer_days_active"] = (pd.Timestamp.today() - df["signup_date"]).dt.days
#------age-----
current_year = datetime.today().year
df["age"] = current_year - df["birth_year"]
```

## ok so now we clip the outliers 

```python
outlier_counts = {}

# Loop through numerical features
for col in new_df.columns:
    Q1 = new_df[col].quantile(0.25)
    Q3 = new_df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    new_df[col] = new_df[col].clip(lower=lower_bound, upper=upper_bound)

    
    
    # Count outliers again
    outliers = df[(new_df[col] < lower_bound) | (new_df[col] > upper_bound)][col]
    outlier_counts[col] = outliers.count()

# Convert to DataFrame for better view
outlier_summary = pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Outlier Count'])
print(outlier_summary)
```

## Ok now we're done with the preprocessing part now to the clustering 

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import numpy as np

# X_scaled = your scaled data (NumPy array or DataFrame)

eps_values = np.linspace(0.1, 1.0, 8)  # 8 eps values from 0.1 to 1.0
min_samples_values = range(3, 11)      # min_samples from 3 to 10

print("eps\tmin_samples\tsilhouette_score\tDBI\tCH_score\t#clusters")

# Track best metrics
best_silhouette = -1
best_silhouette_params = None

best_dbi = float('inf')
best_dbi_params = None

best_ch = -1
best_ch_params = None

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)
        
        # Skip if all points are noise or only one cluster
        if len(set(labels)) <= 1 or (len(set(labels)) == 2 and -1 in labels):
            continue
        
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        
        sil_score = silhouette_score(X_scaled, labels)
        dbi_score = davies_bouldin_score(X_scaled, labels)
        ch_score = calinski_harabasz_score(X_scaled, labels)
        
        print(f"{eps:.2f}\t{min_samples}\t{sil_score:.4f}\t{dbi_score:.4f}\t{ch_score:.2f}\t{n_clusters}")
        
        # Update best scores
        if sil_score > best_silhouette:
            best_silhouette = sil_score
            best_silhouette_params = (eps, min_samples, n_clusters)
        
        if dbi_score < best_dbi:
            best_dbi = dbi_score
            best_dbi_params = (eps, min_samples, n_clusters)
        
        if ch_score > best_ch:
            best_ch = ch_score
            best_ch_params = (eps, min_samples, n_clusters)

# Print results
print("\nBest Silhouette Score:", best_silhouette, "with eps =", best_silhouette_params[0],
      "min_samples =", best_silhouette_params[1], "#clusters =", best_silhouette_params[2])

print("Best Davies-Bouldin Index:", best_dbi, "with eps =", best_dbi_params[0],
      "min_samples =", best_dbi_params[1], "#clusters =", best_dbi_params[2])

print("Best Calinski-Harabasz Score:", best_ch, "with eps =", best_ch_params[0],
      "min_samples =", best_ch_params[1], "#clusters =", best_ch_params[2])
```

```python
from sklearn.cluster import KMeans



cluster_range = range(2, 11)  # 

print("n_clusters\tSilhouette\tDBI\tCH_score\tInertia")

# Track best metrics
best_silhouette = -1
best_silhouette_k = None

best_dbi = float('inf')
best_dbi_k = None

best_ch = -1
best_ch_k = None
inertia_values = []

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    
    sil_score = silhouette_score(X_scaled, labels)
    dbi_score = davies_bouldin_score(X_scaled, labels)
    ch_score = calinski_harabasz_score(X_scaled, labels)
    inertia = kmeans.inertia_
    inertia_values.append(inertia)
    
    print(f"{k}\t{sil_score:.4f}\t{dbi_score:.4f}\t{ch_score:.2f}\t{inertia:.2f}")
    
 
    if sil_score > best_silhouette:
        best_silhouette = sil_score
        best_silhouette_k = k
        
    if dbi_score < best_dbi:
        best_dbi = dbi_score
        best_dbi_k = k
        
    if ch_score > best_ch:
        best_ch = ch_score
        best_ch_k = k


print("\nBest Silhouette Score:", best_silhouette, "with n_clusters =", best_silhouette_k)
print("Best Davies-Bouldin Index:", best_dbi, "with n_clusters =", best_dbi_k)
print("Best Calinski-Harabasz Score:", best_ch, "with n_clusters =", best_ch_k)
```

# Customer Cluster Labels

## Cluster 0 â€” Active but low spenders
- **Top feature(s):** customer_days_active (high)  
- **Other features:** low annual_income & low total_spending  
- **Summary:** Many days active, but relatively low purchases/spend.  
- **Strategy:** Engagement campaigns that upsell or convert activity â†’ spend.

## Cluster 1 â€” Frequent buyers, highly engaged (older)
- **Top feature(s):** num_purchases (highest), customer_days_active (highest), age (2nd highest)  
- **Summary:** Very active customers who buy often; skew older.  
- **Strategy:** Loyalty/repeat-purchase offers, premium products for mature segment.

## Cluster 2 â€” Older, educated, larger families
- **Top feature(s):** age (highest), education_level (highest), family_size (high)  
- **Summary:** Oldest cohort, relatively well-educated, and larger households.  
- **Strategy:** Family bundles, products for multi-person households.

## Cluster 3 â€” High-income campaign responders & spenders
- **Top feature(s):** total_spending (high), annual_income (2nd highest), accepted_campaign_count (high)  
- **Summary:** Responsive to campaigns, high spenders with good income.  
- **Strategy:** Targeted promotions â€” premium/upscale offers and exclusive campaigns.

## Cluster 4 â€” Top-tier wealthy & educated spenders
- **Top feature(s):** annual_income (highest), total_spending (highest), education_level (high), num_purchases (high), accepted_campaign_count (high)  
- **Summary:** Highest income and spending, educated, buys a lot â€” your best customers.  
- **Strategy:** VIP programs, high-ARPU retention strategies, premium product lines.

## Cluster 5 â€” Large families but low engagement & spend
- **Top feature(s):** family_size (2nd highest), but low on almost everything else (spend, purchases, activity)  
- **Summary:** Bigger households but low purchasing and engagement.  
- **Strategy:** Affordability bundles, family-targeted promos to increase uptake.


```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_predict
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings("ignore")  

# -----------------------------
# Prepare X and y
# -----------------------------
X = new_df.drop('Cluster', axis=1)
y = new_df['Cluster']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# -----------------------------
# Models
# -----------------------------
models = {
    'SVM': SVC(kernel='rbf', random_state=42),
    'LogisticRegression': LogisticRegression(solver='lbfgs', max_iter=1000, random_state=42),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(eval_metric='mlogloss', random_state=42)
}


for name, model in models.items():
   
    y_pred = cross_val_predict(model, X_scaled, y, cv=5)
    
    acc = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred, average='macro')
    rec = recall_score(y, y_pred, average='macro')
    f1 = f1_score(y, y_pred, average='macro')
    cm = confusion_matrix(y, y_pred)
    
    print(f"\n{name} Metrics:")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision (macro): {prec:.4f}")
    print(f"Recall (macro): {rec:.4f}")
    print(f"F1-score (macro): {f1:.4f}")
    print("Confusion Matrix:")
    print(cm)
```
